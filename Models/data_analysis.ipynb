```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
max_scores = []
```

```python
models = ["model1", "model2", "model3", "model4", "model5"]
```

```python
m_1 = pd.read_csv('model1_history.csv')
max_scores.append(m_1['Best Score'].max())
```

```python
m_2 = pd.read_csv('model2_history.csv')
max_scores.append(m_2['Best Score'].max())
```

```python
m_3 = pd.read_csv('model3_history.csv')
max_scores.append(m_3['Best Score'].max())
```

```python
m_4 = pd.read_csv('model4_history.csv')
max_scores.append(m_4['Best Score'].max())
```

```python
m_5 = pd.read_csv('model5_history.csv')
max_scores.append(m_5['Best Score'].max())
```

```python
max_scores
```

```python
sns.barplot(x=models, y=max_scores, hue=models)
#plt.plot(models, max_scores, linestyle='-')
plt.title('Comparing Models')
plt.xlabel('Model')
plt.ylabel('Score')
```

```python
optimal_1 = m_1.iloc[m_1['Best Score'].idxmax()].drop(['Generation', 'Best Score']).values.tolist()
```

```python
optimal_2 = m_2.iloc[m_2['Best Score'].idxmax()].drop(['Generation', 'Best Score']).values.tolist()
```

```python
optimal_3 = m_3.iloc[m_3['Best Score'].idxmax()].drop(['Generation', 'Best Score']).values.tolist()
```

```python
optimal_4 = m_4.iloc[m_4['Best Score'].idxmax()].drop(['Generation', 'Best Score']).values.tolist()
```

```python
optimal_5 = m_5.iloc[m_5['Best Score'].idxmax()].drop(['Generation', 'Best Score']).values.tolist()
```

```python
optimal_4
```

```python
best_score_progress = []
for data in [m_1, m_2, m_3, m_4, m_5]:
    best_score_progress.append(data['Best Score'].values.tolist())

plt.figure(figsize=(10, 6))
for i in range(5):
    plt.plot(best_score_progress[i], label=models[i])

plt.title('Best Score Progress')
plt.xlabel('Generation')
plt.ylabel('Score')
plt.legend()
plt.show()
```
